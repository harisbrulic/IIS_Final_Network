```python
import great_expectations as gx

context = gx.get_context()

try:
    context.delete_datasource("cyber_data")
except:
    pass

datasource = context.sources.add_pandas_filesystem(
    name="cyber_data",
    base_directory="../data"
)

normal_asset = datasource.add_csv_asset(
    name="normal_data",
    batching_regex=r"preprocessed\\normal\.csv"
)

attacks_asset = datasource.add_csv_asset(
    name="attacks_data",
    batching_regex=r"preprocessed\\attacks\.csv"
)

threats_asset = datasource.add_csv_asset(
    name="threats_data",
    batching_regex=r"preprocessed\\threats\.csv"
)

print("Done")
```

```python
context.add_or_update_expectation_suite("normal_suite")

context.add_or_update_expectation_suite("attacks_suite")

context.add_or_update_expectation_suite("threats_suite")

print("Done")
```

```python
asset = context.get_datasource("cyber_data").get_asset("normal_data")
batch_request = asset.build_batch_request()

validator = context.get_validator(
    batch_request=batch_request,
    expectation_suite_name="normal_suite"
)

result = context.assistants.onboarding.run(
    validator=validator,
    exclude_column_names=[]
)

context.save_expectation_suite(
    expectation_suite=result.get_expectation_suite(),
    expectation_suite_name="normal_suite"
)

print("Done")
```

```python
asset = context.get_datasource("cyber_data").get_asset("attacks_data")
batch_request = asset.build_batch_request()

validator = context.get_validator(
    batch_request=batch_request,
    expectation_suite_name="attacks_suite"
)

result = context.assistants.onboarding.run(
    validator=validator,
    exclude_column_names=[]
)

context.save_expectation_suite(
    expectation_suite=result.get_expectation_suite(),
    expectation_suite_name="attacks_suite"
)

print("Done")
```

```python
asset = context.get_datasource("cyber_data").get_asset("threats_data")
batch_request = asset.build_batch_request()

validator = context.get_validator(
    batch_request=batch_request,
    expectation_suite_name="threats_suite"
)

validator.expect_table_row_count_to_be_between(min_value=1)
validator.expect_column_values_to_not_be_null("ip")
validator.expect_column_values_to_be_between(
    "confidence", min_value=0, max_value=100
)

validator.save_expectation_suite()

print("Done")
```

```python
normal_asset = context.get_datasource("cyber_data").get_asset("normal_data")
normal_batch = normal_asset.build_batch_request()

context.add_or_update_checkpoint(
    name="normal_checkpoint",
    validations=[{
        "batch_request": normal_batch,
        "expectation_suite_name": "normal_suite"
    }]
)
print("Done")
```

```python
attacks_asset = context.get_datasource("cyber_data").get_asset("attacks_data")
attacks_batch = attacks_asset.build_batch_request()

context.add_or_update_checkpoint(
    name="attacks_checkpoint",
    validations=[{
        "batch_request": attacks_batch,
        "expectation_suite_name": "attacks_suite"
    }]
)
print("Done")
```

```python
threats_asset = context.get_datasource("cyber_data").get_asset("threats_data")
threats_batch = threats_asset.build_batch_request()

context.add_or_update_checkpoint(
    name="threats_checkpoint",
    validations=[{
        "batch_request": threats_batch,
        "expectation_suite_name": "threats_suite"
    }]
)
print("Done")
```

```python
for checkpoint_name in ["normal_checkpoint", "attacks_checkpoint", "threats_checkpoint"]:
    checkpoint = context.get_checkpoint(checkpoint_name)
    result = checkpoint.run()
    if result["success"]:
        print(f"{checkpoint_name}")
    else:
        print(f"{checkpoint_name}")

context.build_data_docs()
context.open_data_docs()
```
